```python
class Demo:
  def add(self,a=0,b=0,c=0):
    print(a+b+c)

obj = Demo()
obj.add()
obj.add(10,20)
obj.add(1,2,23)
obj.add("K", "a")
```

```python
class Demo2:
  def add(self, *args):
    print(args)

obj = Demo2()
obj.add(1,2,3,4,4,5)
obj.add("K", "a")
```

```python
#Function Overriding
class A:
  def display(self):
    print("This is class A")
class B(A):
  def display(self):
    print("This is class B")

obj = B()
obj.display()
```

```python
class Animal:
    def sound():
        pass

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

class Cow(Animal):
    def sound(self):
        print("Moo")

class Lion(Animal):
    def sound(self):
        print("Roar")

class Snake(Animal):
    def sound(self):
        print("Hiss")

class Duck(Animal):
    def sound(self):
        print("Quack")

class Main:
    def main():
        option = input()

        if option == 'Dog':
            obj = Dog()
        elif option == 'Cat':
            obj = Cat()
        elif option == 'Cow':
            obj = Cow()
        elif option == 'Lion':
            obj = Lion()
        elif option == 'Snake':
            obj = Snake()
        elif option == 'Duck':
            obj = Duck()
        else:
            print("Invalid Input")
            return

        obj.sound()

Main.main()
```
