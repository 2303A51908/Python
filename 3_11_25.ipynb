```python
import math
#1. round(x): Round to the nearest value
a = 2
b = 3
print(b/a)
print(round(b/a))

#2. floor(x): the largest integer less than or equal to x

print(math.floor(b/a))

#3.ceil(x): the smallest integer greather than or equal to x

print(math.ceil(b/a))

#4.pow(x,y): x raised to the power of y

print(pow(a,b))

#5. abs(x): absolute value of x
print("abs:", abs(-2.34))

#sqrt(x): square root of x
x = 25
print(int(math.sqrt(x)))

#caluclate the factorial of 5 (5! = 5*4*3*2*1)

print("Fact of 5 :", math.factorial(5))

#Exponential
#caluclate e^2 eucler number e =2.718
print("e^2=", math.exp(2))

#logorithm
print("Log base 10 of 100:", math.log10(100))

#Trigonometry

print("sin(90):", math.sin(math.radians(90)))

print("tan(0):", math.tan(0))

#Convert degrees to radians :- pi radians = 180 degrees
print("Degrees of pi radians:", math.degrees(math.pi))

#convert radians to degree :- 180 dergree = pi radians
print("Radians of 180 degrees:", math.radians(180))

#GCD(greatest common divisor)
print("GCD of 12 and 15:", math.gcd(12, 15))

#Integer square root
#integer square root of 20:- largest integer <= sqrt (20)=4
print("Integer square root of 20:", math.isqrt(20))
```
