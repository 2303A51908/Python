```python
#Write a python program to count no.of students with the grade A in the following tuple

grade = ('C', 'A', 'D', 'B', 'A', 'A')
student = grade.count('A')
print(student)

#Using for loop
grade = ('C', 'A', 'D', 'B', 'A', 'A')
count = 0
for grade in grade:
  if grade == 'A':
    count += 1
print(count)
```

```python
#Count the No.of Vowels in the given string
str = input("Enter a String: ")
vowels = "aeiouAEIOU"
count = 0
for ch in str:
  if ch in vowels:
    count += 1
print(count)
```

```python
# Find the factorial of first n numbers using loop
n = int(input("Enter a Number: "))
fact = 1
for i in range(1, n+1):
  fact = fact*i
  print(fact)
print("5! =", fact)
```
